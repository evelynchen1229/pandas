```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
# unsupervised learning means we are not trying to predict any output, but just patterns in the data
# K-Means Clustering we need to specify the number of clusters you want for the data to be grouped in
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data
```

```python
data[0].shape # 200 samples 2 features
```

```python
plt.scatter(data[0][:,0],data[0][:,1]) # plot data[0] all the rows 1 col vs 2col
# two blobs here, need to grab the second item in the tuple
```

```python
data[1]
# second item in the tuple; this is actually the clusters that it belongs to (number is the number from the centers and there are four centers)
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow') 
# there are four blobs/clusters here
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
#need to know how many clusters
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
#predicted labels
#use it to try to find labels in your data instead of actually predicting them
```

```python
data[1]
#correct labels
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))
#sharey: share axis

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')#color based on where the algorithms thinks are

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
#color between the two plots doesn't mean anything, not referencing back to each other
#notice below, though, there's some noise in original but the algorithms couldn't decide what to do with them , so assign them to the red cluster centroid (yellow as well)
```
